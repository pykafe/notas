```python
import pandas as pd # pandas funsaun ida husi python atu halo analiza
import numpy as np # numpy funsaun ida husi python atu halo analiza valor numerikal/numeric sira.
import matplotlib.pyplot as plt # matplotlib funsaun ida husi python atu halo hatudu vizualizar ka hatudu grafiku sira.
import xlsxwriter # pandas funsaun ida husi python atu halo hakerek dile excel iha pandas


rezultadu = []
bob = pd.read_excel('valor.xlsx')
for i in bob["valor"]:
    if i >= 50:
        rezultadu.append("aprovadu")
    else:
        rezultadu.append("reprovadu")
        
bob['Rezultadu'] = rezultadu

writer = pd.ExcelWriter('valor.xlsx', engine='xlsxwriter') # xlsxwrite funsaun ida husi pandas nian atu kria valor no rai iha file excel nian.

bob.to_excel(writer, sheet_name='Sheet1') # Export file husi pandas nia ba sheet iha excel nian.

writer.save()# rai file nebe hakerek iha pandas ba excel
bob
```

```python
bob.plot(kind="bar", title="test")
plt.show()
```

```python
import pandas as pd # pandas funsaun ida husi python atu halo analiza
import numpy as np # numpy funsaun ida husi python atu halo analiza valor numerikal/numeric sira.
import matplotlib.pyplot as plt # matplotlib funsaun ida husi python atu halo hatudu vizualizar ka hatudu grafiku sira.
import xlsxwriter # pandas funsaun ida husi python atu halo hakerek dile excel iha pandas

rezultadu = []
df = pd.read_excel('valor_estudantes.xlsx')
df["total"]=df["tri1"] + df["tri2"] + df["tri3"]
for i in df["total"]:
    if i / 3 >= 50:
        rezultadu.append("aprovadu")
    elif i / 3 < 50: 
        rezultadu.append("reprovadu")
    else:
        rezultadu.append('Failla')
df['Rezultadu'] = rezultadu
writer = pd.ExcelWriter('valor_estudantes.xlsx', engine='xlsxwriter') # xlsxwrite funsaun ida husi pandas nian atu kria valor no rai iha file excel nian.
df.to_excel(writer, sheet_name='Sheet1') # Export file husi pandas nia ba sheet iha excel nian.
writer.save()# rai file nebe hakerek iha pandas ba excel
df
```
